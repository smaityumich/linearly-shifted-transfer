```python
import pandas as pd

def highlight(s):
    if s.pvalue < 0.05:
        return ['background-color: yellow']*7
    else:
        return ['background-color: white']*7
```

```python
white_interaction = pd.read_csv('white.interaction.csv')    
white_interaction.style.apply(highlight, axis=1)
```

```python
white_main = pd.read_csv('white.main.csv')    
white_main.style.apply(highlight, axis=1)
```

```python
asian_interaction = pd.read_csv('asian.interaction.csv')    
asian_interaction.style.apply(highlight, axis=1)
```

```python
asian_main = pd.read_csv('asian.main.csv')    
asian_main.style.apply(highlight, axis=1)
```

```python
asian_offset = pd.read_csv('asian.offset.csv')    
asian_offset.style.apply(highlight, axis=1)
```
