```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import json, os, glob, re

import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = [
    r'\usepackage{amsmath}',
    r'\usepackage{amssymb}']
```

```python
path = 'summaries_part/'

list_dir = []

for file in glob.glob(path + '*.json'):
    with open(file) as f:
        data = json.load(f)
        
    new_dict = data['par']
    for test in ['source', 'target']:
        for model, accuracy in zip(data['accuracy']['name'], data['accuracy'][test]):
            if accuracy == 'NA':
                accuracy = 'NaN'
            new_dict[(test, model)] = accuracy
            
    list_dir.append(new_dict)

combined = []
for file in glob.glob(path + '*.json'):
    with open(file) as f:
        data = json.load(f)
        combined.append(data)
        
with open('summary/summary_all_link.json', 'w') as f:
    json.dump(combined, f)
    
# with open('summary/summary_logit.json') as f:
#     combined = json.load(f)

# list_dir = []

# for data in combined:
#     new_dict = data['par']
#     for test in ['source', 'target']:
#         for model, accuracy in zip(data['accuracy']['name'], data['accuracy'][test]):
#             if accuracy == 'NA':
#                 accuracy = 'NaN'
#             new_dict[(test, model)] = accuracy
            
#     list_dir.append(new_dict)    
```

```python
df = pd.DataFrame(list_dir)
```

```python
defaults = {'m': 2000, 'n': 50, 'delta': 2.0, 'sigma': 1.0, 'd': 5}
key_labels = ['$m$', '$n$', r'$\delta$', r'$\sigma$', '$d$']
cols = ['r', 'b', 'g', 'k', 'y', 'cyan']
ltys = ['-', '--', '-', '--', '-.', ':']
mks = ['o', '+', 'o', '+', 'x', '>']


fig, ax = plt.subplots(2, 3, figsize = (15, 10))

variables = dict()
for _ in range(len(df.columns)):
    index = df.columns[_]
    if index[0] == 'target':
        variables[index] = ['mean', 'std']
        

        
for i, key, kl in zip(range(6), defaults.keys(), key_labels):
    df_part = df
    for key_default in defaults.keys():
        if key_default != key:
            df_part = df_part.loc[df_part[key_default] == defaults[key_default]]
            
    a = df_part.dtypes == 'object'
    a = list(a.loc[a == True].index)
    for col in a:
        df_part = df_part.astype({col: 'float'})
            
    summary = df_part.groupby([key], as_index=False).agg(variables)
    x = summary[key]
    
    for index, col, lty, mk in zip(variables, cols, ltys, mks):
        mean = summary[index]['mean']
        std = summary[index]['std']
        u = i//3
        v = i%3
        ax[u][v].errorbar(x, mean, std, color = col, linestyle = lty, marker = mk)
        ax[u][v].set_xlabel(kl, fontsize = 20)
        if key != 'd':
            ax[u][v].set_xscale('log')
        
        
        
names = []
lines = []
for index, col, lty, mk in zip(variables, cols, ltys, mks):
    names.append(index[1])
    lines.append(Line2D([0], [0], color = col, linestyle = lty, marker = mk))
    
ax[1][2].legend(lines, names, loc = 'center', fontsize = 15)
ax[1][2].get_xaxis().set_visible(False)
ax[1][2].get_yaxis().set_visible(False)  
fig.savefig('plot/plot_cloglog.pdf', bbox_inchs = 'tight')
```

```python
path = 'summaries_part/'

list_dir = []

for file in glob.glob(path + '*.json'):
    with open(file) as f:
        data = json.load(f)
        
    new_dict = data['par']
    for test in ['source', 'target']:
        for model, accuracy in zip(data['accuracy']['name'], data['accuracy'][test]):
            if accuracy == 'NA':
                accuracy = 'NaN'
            new_dict[(test, model)] = accuracy
            
    list_dir.append(new_dict)
    
    
df = pd.DataFrame(list_dir)

variables = dict()
variables[('target', 'transfer')] = ['mean', 'std']
variables[('target', 'val.full')] = ['mean', 'std']

summary = df.groupby(['link','d'], as_index=False).agg(variables)

markers = ['x', 'o', '+', '<']
colors = ['g', 'b', 'r', 'm']
links = ['logit', 'probit', 'cauchit', 'cloglog']
ltys = ['-', '--', '-.', ':']
lines = []

for marker, color, link, lty in zip(markers, colors, links, ltys):
    summary_part = summary.loc[summary['link'] == link]
    x = summary_part['d']
    mean = summary_part[('target', 'transfer')]['mean']
    std = summary_part[('target', 'transfer')]['std']
    
    plt.errorbar(x, mean, std, marker = marker, linestyle = lty, color = color)
    lines.append(Line2D([0], [0], marker = marker, linestyle = lty, color = color))
    
labels = links

# summary_part = summary.loc[summary['link'] == 'logit']
# x = summary_part['d']
# mean = summary_part[('target', 'val.full')]['mean']
# std = summary_part[('target', 'val.full')]['std']
# plt.errorbar(x, mean, std, marker = '*', linestyle = '-', color = 'k')
# lines.append(Line2D([0], [0], marker = '*', linestyle = '-', color = 'k'))
# labels = labels + ['best achievable']

plt.legend(lines, labels, fontsize = 12, title = 'data link', title_fontsize = 20)
plt.xlabel('$d$', fontsize = 15)
plt.ylabel('accuracy on $Q$', fontsize = 15)
plt.savefig('plot/link_d.pdf', bbox_inchs = 'tight')
```
